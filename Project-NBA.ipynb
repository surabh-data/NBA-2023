```python
from nba_api.stats.static import teams
# https://pypi.org/project/nba-api/
nba = teams.get_teams()
nba
```

Checking and Extracting keys from the dictionary : Heading of the Columns

```python
nba_keys = list(nba[0].keys())
nba_keys
        
```

Extracting the Dictionary values in the list from # nba list to a new list in the following code.

```python
response= [nba_keys]
def rows(team):
    global response
    for a in range(len(team)):
        nba_keys1 = []
        for i in team[a]:
            nba_keys1.append(team[a][i])
        response.append(nba_keys1)
    return response
nba_team_list= rows(nba)
nba_team_list
```

Checking the value to write over the new file must be string data type.
String data type value can be written in csv file.

```python
for rows in range(len(nba_team_list)):
        for value in range(len(nba_team_list[rows])):
            point1 = f"{nba_team_list[rows][value]},"
# (type(point1))
```
